```python
import matplotlib.pyplot as plt
import seaborn 
import pandas as pd
```

```python
# Diabetes subset CB NoWeights
model_data = {'Epoch': [1, 2, 3, 4], 'Diabetes Training Loss': [0.149, 0.130, 0.116, 0.091],
              'Diabetes Validation Loss': [0.138, 0.137, 0.141, 0.160],
              'Train Accuracy': [0.930, 0.931, 0.940, 0.940], 
              'Validation Accuracy': [0.931, 0.930, 0.927, 0.914],'Hypertension Training Loss': [0.276, 0.254, 0.237, 0.226],
              'Hypertension Validation Loss': [0.270, 0.251, 0.242, 0.242],}
model_df = pd.DataFrame(data=model_data)
model_df.set_index('Epoch')
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

ax1.set(xlabel='Epochs', ylabel='Cross Entropy Loss')
ax2.set(xlabel='Epochs', ylabel='Cross Entropy Loss')
fig.suptitle('BioDischargeSummaryBERT with Weights: Loss over Epochs')
seaborn.lineplot(data=model_df[['Diabetes Training Loss', 'Diabetes Validation Loss']], ax=ax1)
seaborn.lineplot(data=model_df[['Hypertension Training Loss', 'Hypertension Validation Loss']], ax=ax2)
```
